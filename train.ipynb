```python
import json
train_csv = '/content/Multi-Label-Text-Classification-for-Chinese/mltc/dataset/train.csv'
data_list = []
with open('/content/final_all_data/exercise_contest/data_train.json', encoding='utf-8') as f:
  data = f.read().strip().split('\n')
for d in data:
  tmp = json.loads(d)
  data_list.append([tmp["fact"], tmp["meta"]["accusation"]])

#print(data_list[1664])

with open(train_csv,'w', encoding='utf-8') as f:
  f.write('fact,meta\n')
  for d in data_list:
    d[0] = d[0].replace('"','”')
    f.write(f'"{d[0]}",{"----".join(d[1]).strip()}\n')



```
